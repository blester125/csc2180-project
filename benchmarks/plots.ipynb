```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.color_palette("tab10")
sns.set_theme()

%matplotlib inline
```

```python
import z3ml
```

```python
with open("results.json") as f:
    data = json.load(f)
```

```python
df = pd.DataFrame(data)
```

```python
df = df[df.processor.str.contains("AMD")]

# Remove Outlier
df = df[~((df.num_classes == 4) & (df.num_features == 3) & (df.num_samples == 10000) & (df.seed == 1760814101))]
```

# Samples

```python
d = df[(df.num_features == 3) & (df.num_classes == 4)]
```

```python
sns.lineplot(data=d[d.one_vs_one == False], x="num_samples", y="solve_time");
```

```python
sns.lineplot(data=d[d.one_vs_one == True], x="num_samples", y="solve_time", color="tab:orange");
```

```python
sns.lineplot(data=d, x="num_samples", y="solve_time", hue="one_vs_one");
```
